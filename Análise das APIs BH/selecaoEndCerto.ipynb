```python
import pandas as pd
```

```python
geoGold = pd.read_csv('geoGold_erro.csv')
geoGold.head()
```

```python
requests_id = geoGold['request_id'].unique()
requests_id = list(requests_id)
```

```python
result_df = geoGold.iloc[0:0]

for request in requests_id:
    aux = geoGold[geoGold['request_id'] == request]
    min_row = aux.loc[aux['erro'].idxmin()]
    result_df = result_df.append(min_row, ignore_index=True)

    
```

```python
result_df.head()
```

```python
result_df.columns
```

```python
result = result_df[['number_x', 'CEP', 'street', 'bairro', 'city', 'state',
                    'geoapi_id', 'request_id','lat', 'long', 'accuracy', 
                    'district', 'public_place','erro']]
```
