```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
original = pd.read_csv('BaseBHtransformada.csv')
amostra = pd.read_csv('amostraBH.csv')
```

```python
plt.plot(original['LONGITUDE'], original['LATITUDE'], '.', markersize=0.1)

plt.title('Complete Belo Horizonte Address Database')  # Adiciona um ttulo ao grfico
plt.xlabel('Longitude')  # Adiciona legenda ao eixo x
plt.ylabel('Latitude')  # Adiciona legenda ao eixo y

plt.show()
```

```python
plt.plot(amostra['LONGITUDE'], amostra['LATITUDE'], '.', markersize=0.1)

plt.title('Sampled Belo Horizonte Address Database')  # Adiciona um ttulo ao grfico
plt.xlabel('Longitude')  # Adiciona legenda ao eixo x
plt.ylabel('Latitude')  # Adiciona legenda ao eixo y

plt.show()
```
