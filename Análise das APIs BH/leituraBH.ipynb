Leitura dos dados Ouro

```python
import pandas as pd
```

```python
amostra = pd.read_csv('..\csv\amostraBH.csv')
padronizado = pd.read_csv('..\csv\amostraBHPadronizada.csv')
amostra_lat_long = amostra[["LATITUDE", "LONGITUDE"]]
original = padronizado.copy()

# Adicionar as colunas "LATITUDE" e "LONGITUDE" do dataframe "amostra" ao dataframe "original"
original[["lonGold", "latGold"]] = amostra_lat_long
if 'Unnamed: 0' in original.columns:
    original.drop('Unnamed: 0', axis=1, inplace=True)

original['id'] = range(2767600, 2767600 + len(original))
original.head()
```

```python
geo = pd.read_csv('..\csv\respostaBh.csv')
```

Junção dos dados geocodificados com os dados ouro

```python
result_df = original.merge(geo, how='inner', left_on='id', right_on='request_id')
result_df.head(50)
```

```python
aux = result_df.drop_duplicates(subset=['request_id', 'geoapi_id'])
```

```python
result_df = aux
```
